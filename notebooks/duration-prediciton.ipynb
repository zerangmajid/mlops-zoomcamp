```python
# Calculate trip duration as a timedelta between dropoff and pickup
df['duration'] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime

# Convert duration from timedelta to total minutes (as float)
df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)
```
